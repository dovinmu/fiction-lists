```python
from bs4 import BeautifulSoup
import re
import requests
import time
```

```python
url = 'https://firesidefiction.com/magazine'
```

```python
with open('Fireside Magazine.html') as f:
    soup = BeautifulSoup(f.read())
```

```python
stories = []
for issue in soup.findAll('div', attrs={'class':'issue-container'}):
    date = issue.previous_sibling.previous_sibling.text
    if '2021' not in date:
        continue
    for entry in issue.findAll('h3'):
        if 'short story' in entry.text:
#             print(entry.text)
            href = 'https://firesidefiction.com'+entry.find('a').get('href')
            title, author = entry.text.split(', a short story by')
            stories.append((title, author.strip(), href))
stories
```

```python
with open('fireside-2021-shorts.md', 'w') as f:
    for title, author, href in sorted(list(stories), key=lambda el: el[0]):
        f.write(f"{title} by {author}: {href}\n")
```
